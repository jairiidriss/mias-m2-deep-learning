# Supervised Learning Algorithms: k-Nearest Neighbors (k-NN)

<b>k-Nearest Neighbors (k-NN)</b> is a supervised machine learning algorithm used for both classification and regression tasks. It's a simple and intuitive algorithm that relies on the idea of finding the "nearest" data points in the training set to a given test point and using those neighbors to make a prediction.
